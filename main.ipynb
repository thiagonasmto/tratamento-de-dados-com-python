```python
dados = carregar_dados("./heart_attack_prediction_dataset.csv")
dados.head()
```

```python
df = dados.copy()
df[['Systolic Pressure','Asystolic Pressure']] = dados['Blood Pressure'].str.split('/', expand=True).astype(int)
df = df.drop('Blood Pressure', axis=1)
df.head(10)
```

```python
diet_map = {
    'Average':   0.5,
    'Unhealthy': 0,
    'Healthy':   1
}


df['Diet'] = dados['Diet'].replace(diet_map).astype(float)

sex_map = {
    'Male':   0,
    'Female': 1,
}

df['Sex'] = dados['Sex'].replace(sex_map).astype(int)

df.head(10)
```
