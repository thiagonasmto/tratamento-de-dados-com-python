```python
df = dados.copy()
df[['Systolic Pressure','Asystolic Pressure']] = dados['Blood Pressure'].str.split('/', expand=True).astype(int)
df = df.drop('Blood Pressure', axis=1)
df.head(10)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import os

# Carregando dataset
caminho_arquivo = 'dados_normalizados.csv'
df = pd.read_csv(caminho_arquivo)
df = df.drop(['Patient ID','Country','Continent','Hemisphere'], axis=1)

# Escolhendo o número de clusters
n_clusters = 3

# Aplicando K-means
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(df)

# Adicionando os rótulos dos clusters ao DataFrame
df['Cluster'] = kmeans.labels_

# Criando diretório para salvar os arquivos CSV (se não existir)
output_dir = 'clusters_output'
os.makedirs(output_dir, exist_ok=True)

# Salvando cada cluster em um arquivo CSV
for cluster_num in range(n_clusters):
    cluster_data = df[df['Cluster'] == cluster_num]
    filename = f'cluster_{cluster_num}.csv'
    filepath = os.path.join(output_dir, filename)
    cluster_data.to_csv(filepath, index=False)

print(f"Arquivos CSV salvos na pasta '{output_dir}'")
```
